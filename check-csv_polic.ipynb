```python
df["address"] = df["ObjectAddress"].str.extract(r"Address:(.*)")
```

```python
df = df.drop(index=0).reset_index(drop=True)
```

```python
df.head(10)
```

```python
import re

def to_point_wkt(value):
    lon, lat = re.findall(r"[-\d.]+", value)[:2]
    return f"POINT({lon} {lat})"

df["location"] = df["geoData"].apply(to_point_wkt)
```

```python
df.head(10)
```

```python
df.drop(columns=["ObjectAddress"], inplace=True)
df.drop(columns=["geoData"], inplace=True)
df.drop(columns=["geometry"], inplace=True)
df.head(10) 
```

```python
df_unique = df.drop_duplicates(subset=["ShortName"], keep="first")
```

```python
# df_unique.info()
# df_unique.describe()
df['ShortName'].value_counts()
# df_unique['address'].value_counts().head(10)
```

```python
df_unique = df[df["ShortName"].str.len() > 0]
```

```python
df_unique = df_unique[df_unique["ShortName"].notna()]
df_unique.info()
```

```python
df_unique = df_unique[~df_unique["ShortName"].str.contains("none", case=False, na=False)]
df_unique.info()
```

```python
df_unique['address'].value_counts()
```

```python
last = df["address"].str.split(",").str[-1].str.strip()
has_number = last.str.contains(r"\d", na=False, regex=True)

df_unique = df_unique[has_number]  # оставить только строки, где в «хвосте» адреса есть цифра
df_unique.info()
```

```python
df_unique['address'].value_counts()
```

```python
df_unique["address"] = df_unique["address"].str.replace(
    r"(\d+)\s*([кКkK])\s*(\d+)",
    r"\1 корпус \3",
    regex=True,
)
```

```python
df_unique['address'].value_counts()
```

```python
df_unique.head(1)
```

```python
df_unique = df_unique.rename(columns={
    "Category": "category",
    "FullName": "fullname",
    "ShortName": "shortname",
    "Address": "address",
})
```

```python
df_unique.to_csv("policlinics_2.csv", index=False, encoding="utf-8")
```
