```python
df_n_b = df[df['building_year'] >= 2025]
```

```python
df_unique = df_n_b.drop_duplicates(subset=["address"], keep="first")
```

```python
# df_unique.info()
# df_unique.describe()
df_unique['address'].value_counts()
# df_unique['address'].value_counts().head(10)
```

```python
df_unique = df_unique[df_unique["address"].str.len() > 0]
```

```python
df_unique = df_unique[df_unique["address"].notna()]
df_unique.info()
```

```python
df_unique = df_unique[~df_unique["address"].str.contains("none", case=False, na=False)]
df_unique.info()
```

```python
df_unique['address'].value_counts()
```

```python
last = df["address"].str.split(",").str[-1].str.strip()
has_number = last.str.contains(r"\d", na=False, regex=True)

df_unique = df_unique[has_number]  # оставить только строки, где в «хвосте» адреса есть цифра
df_unique.info()
```

```python
df_unique['address'].value_counts()
```

```python
df_unique["address"] = df_unique["address"].str.replace(
    r"(\d+)\s*([кКkK])\s*(\d+)",
    r"\1 корпус \3",
    regex=True,
)
```

```python
df_unique['address'].value_counts()
```

```python
df_unique.to_csv("new_buildings.csv", index=False, encoding="utf-8")
```
